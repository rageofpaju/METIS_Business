# I am importing hourly PM data of Seoul from 2016~2021, then keeping only the average values of each hour (removing all the 'by district' data) 

## *Because destricts were written in Korean, it couldn't be imported with read_csv. Therefore, all the rows with districts other than average were removed via excel first 


```python
import pandas as pd
```

```python
pm_1 = pd.read_csv('서울시 대기질 자료 제공_2016-2019_cleaned.csv')
pm_2 = pd.read_csv('서울시 대기질 자료 제공_2020-2021_cleaned.csv')
```

```python
pm_1619 = pm_1.dropna(subset=["Date"])
pm_2021 = pm_2.dropna(subset=["Date"])
```

```python
pm_2021.info()
```

```python
pm_1621 = pd.concat([pm_1619, pm_2021], ignore_index=True)
```

```python
pm_1621.to_csv('pm_16to21.csv', encoding = 'utf-8-sig')
```
